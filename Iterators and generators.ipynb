```python
l = [1,2,3,4]
```

```python
l_iterator = iter(l)
```

```python
for ele in l:
    print(ele)
```

```python
while True:
    try:   
        ele = next(l_iterator)    
        print(ele)
    except StopIteration:
        break
```

```python
# A simple generator 
def simple_gen():
    #print('I am here ')
    yield 'a'
    yield 'b'
    yield 'c'
```

```python
x = simple_gen()
```

```python
def gen_demo1():
    for x in range(5):
        yield x
```

```python
x = gen_demo1()
```

```python
next(x)
```
